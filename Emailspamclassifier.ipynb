```python
import os #navigate operating system
from collections import Counter
```

```python
files=os.listdir('email')
```

```python
emails=[]
for i in files:
    emails.append('email/' + i)
```

```python
emails
```

```python
words=[]
for i in emails:
    f=open(i,encoding='latin-1')
    words.extend(f.read().split(" "))
```

```python
len(words)
```

```python
"Flat42".isalpha()
```

```python
for i in range(len(words)):
    if not words[i].isalpha():
        words[i]=""
```

```python
word_dict=Counter(words)
```

```python
word_dict.most_common(1)
```

```python
del word_dict['']
```

```python
word_dict
```

```python
len(word_dict)
```

```python
new_word_dict=word_dict.most_common(3000)
```

```python
new_word_dict
```

```python
features=[]
labels=[]
for i in emails:
    if 'spam' in i:
        labels.append(1)
    if 'ham'  in i:
        labels.append(0)
    f=open(i,encoding='latin-1')
    text=f.read()
    
    data=[]
    for j in new_word_dict:
        data.append(text.count(j[0]))
    features.append(data)    
```

```python
len(features[0])
```

```python
len(labels)
```

```python
import numpy as np
```

```python
X=np.array(features)
y=np.array(labels)
```

```python
print(X.shape)
```

```python
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
```

```python
clf.fit(X_train,y_train)
```
